```python
# Importing all libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# connecting the google drive
from google.colab import drive
drive.mount('/content/drive')

#creating the empty list to store information about the book
books = []

for i in range(1, 5):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response, 'html.parser')

    # Finding the 'ol' (ordered list) element containing book information
    ol = soup.find('ol')

    # Finding all 'article' elements with the class 'product_pod'
    articles = ol.find_all('article', class_='product_pod')

    # Loop through each book 'article' element
    for article in articles:
        # Finding the 'img' element to extract the book title from its 'alt' attribute
        image = article.find('img')
        title = image.attrs['alt']

        # Finding the 'p' (paragraph) element to extract the star rating from its 'class' attribute
        starTag = article.find('p')
        star = starTag['class'][1]

        # Finding the 'p' element with the class 'price_color' to extract the book price
        price = article.find('p', class_='price_color').text

        # Appending the book information to the 'books' list
        books.append([title, star, price, url])

# Create a DataFrame using pandas
df = pd.DataFrame(books, columns=['Title', 'Star Rating', 'Price', 'Link'])

# Save the DataFrame to a CSV file in Google Drive
csv_file_path = '/content/drive/MyDrive/PREPINSTA DATA ANYLITICS INTERSHIP PROJECTS /WEEK 5 PROJECT /PROJECT /books.csv'
df.to_csv(csv_file_path, index=False)

# Display the DataFrame
df
```
